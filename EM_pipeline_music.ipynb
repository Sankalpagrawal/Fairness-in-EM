Matching

```python
# Sample candidate set
S = em.sample_table(C, 538)
```

```python
# Label S
G = em.label_table(S, 'label')
```

```python
# Load the pre-labeled data
path_G = 'music_labeled_data.csv'
# path_G = em.get_install_path() + os.sep + 'datasets' + os.sep + 'end-to-end' + os.sep + 'restaurants/lbl_restnt_wf1.csv'
G = em.read_csv_metadata(path_G, 
                         key='_id',
                         ltable=A, rtable=B, 
                         fk_ltable='ltable.Sno', fk_rtable='rtable.Sno')
print(len(G))
```

```python
# Split S into development set (I) and evaluation set (J)
IJ = em.split_train_test(G, train_proportion=0.7, random_state=0)
I = IJ['train']
J = IJ['test']
```

```python
# Create a set of ML-matchers
dt = em.DTMatcher(name='DecisionTree', random_state=0)
svm = em.SVMMatcher(name='SVM', random_state=0)
rf = em.RFMatcher(name='RF', random_state=0)
lg = em.LogRegMatcher(name='LogReg', random_state=0)
ln = em.LinRegMatcher(name='LinReg')
nb = em.NBMatcher(name='NaiveBayes')
```

```python
# Generate features
feature_table = em.get_features_for_matching(A, B, validate_inferred_attr_types=False)
```

```python
# List the names of the features generated
feature_table['feature_name']
```

```python
# Convert the I into a set of feature vectors using F
H = em.extract_feature_vecs(I, 
                            feature_table=feature_table, 
                            attrs_after='label',
                            show_progress=False)
```

```python
# Display first few rows
H.head(3)
```

```python
# Select the best ML matcher using CV
result = em.select_matcher([dt, rf, svm, ln, lg, nb], table=H, 
        exclude_attrs=['_id', 'ltable.Sno', 'rtable.Sno', 'label'],
        k=5,
        target_attr='label', metric_to_select_matcher='f1', random_state=0)
result['cv_stats']
```

```python
#  Split feature vectors into train and test
UV = em.split_train_test(H, train_proportion=0.5)
U = UV['train']
V = UV['test']
```

```python
# Debug decision tree using GUI
em.vis_debug_rf(rf, U, V, 
        exclude_attrs=['_id', 'ltable.Sno', 'rtable.Sno', 'label'],
        target_attr='label')
```

```python
match_t = em.get_tokenizers_for_matching()
match_s = em.get_sim_funs_for_matching()
atypes1 = em.get_attr_types(A) # don't need, if atypes1 exists from blocking step
atypes2 = em.get_attr_types(B) # don't need, if atypes2 exists from blocking step
match_c = em.get_attr_corres(A, B)
match_f = em.get_features(A, B, atypes1, atypes2, match_c, match_t, match_s)
```

```python
# def phone_phone_feature(ltuple, rtuple):
#     p1 = ltuple.phone
#     p2 = rtuple.phone
#     p1 = p1.replace('/','-')
#     p1 = p1.replace(' ','')
#     p2 = p2.replace('/','-')
#     p2 = p2.replace(' ','')    
#     if p1 == p2:
#         return 1.0
#     else:
#         return 0.0

def time_feature(ltuple, rtuple):
    p1 = ltuple.Time
    p2 = rtuple.Time
    if p1 == p2:
        return 1.0
    else:
        return 0.0

feature_table = em.get_features_for_matching(A, B)
em.add_blackbox_feature(feature_table, 'time_time_feature', time_feature)
H = em.extract_feature_vecs(I, feature_table=feature_table, attrs_after='label', show_progress=False)

# Select the best ML matcher using CV
result = em.select_matcher([dt, rf, svm, ln, lg, nb], table=H, 
        exclude_attrs=['_id', 'ltable.Sno', 'rtable.Sno', 'label'],
        k=5,
        target_attr='label', metric_to_select_matcher='f1', random_state=0)
result['cv_stats']
```

```python
# Convert J into a set of feature vectors using feature table
L = em.extract_feature_vecs(J, feature_table=feature_table,
                            attrs_after='label', show_progress=False)
```

```python
# Train using feature vectors from I 
dt.fit(table=H, 
       exclude_attrs=['_id', 'ltable.Sno', 'rtable.Sno', 'label'], 
       target_attr='label')
```

```python
# Predict on L 
predictions = dt.predict(table=L, exclude_attrs=['_id', 'ltable.Sno', 'rtable.Sno', 'label'], 
              append=True, target_attr='predicted', inplace=False)
```

```python
# Evaluate the predictions
eval_result = em.eval_matches(predictions, 'label', 'predicted')
em.print_eval_summary(eval_result)
```
